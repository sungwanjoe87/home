```python
import sklearn
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
from sklearn.model_selection import train_test_split
matplotlib.rcParams['font.family'] = 'DejaVu Sans'
matplotlib.rcParams['axes.unicode_minus'] = False
from tensorflow import keras
```

```python
## mnist 파일 불러오기.
## keras mnist 파일이 자동 설치되어 있음.
(train_data,train_label),(test_data,test_label) = \
    keras.datasets.fashion_mnist.load_data()
    
```

```python
## 전처리
train_scaled = train_data/255.0
test_scaled = test_data/255.0
```

```python
### 모델 선언.
model = keras.Sequential()
model.add(keras.layers.Flatten(input_shape=(28,28)))
model.add(keras.layers.Dense(100,activation='relu'))
model.add(keras.layers.Dense(10,activation='softmax'))
```

```python
### 딥러닝 설정.

model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',\
    metrics='accuracy')
```

```python
### 훈련.

history = model.fit(train_scaled,train_label,epochs=20,validation_data=(test_scaled,test_label))
### 에포크 만큼의 [loss:손실율,accuracy]:정확도 출력
print(history.history['loss'])
print(history.history['accuracy'])
```
